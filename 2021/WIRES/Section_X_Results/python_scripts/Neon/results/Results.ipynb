```python
import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
```

---
### Wu and Yang

```python
with open('wuyang.json', encoding='utf-8') as F:
    wuyang = json.loads(F.read())
```

```python
fig, ax = plt.subplots(1, 5, figsize=(20,5))
for i in range(5):
    ax[i]
    ax[i].plot(wuyang["grid"], wuyang["vxc"][i], label=wuyang["basis"][i])
    ax[i].text(0.1,-9, s=f"Homo Value:\n {wuyang['occ_eigs'][i][4]:6f}")
    ax[i].legend()
    ax[i].set_ylim(-14,1)
    ax[i].set_xlim(0.001,9)
    ax[i].set_xscale('log')
    
fig.suptitle("Neon: WuYang Method")
```
