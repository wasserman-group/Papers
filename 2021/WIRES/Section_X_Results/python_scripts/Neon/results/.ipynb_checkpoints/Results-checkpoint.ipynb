```python
import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
```

```python
npoints=5000
x = np.linspace(0,10,npoints)[:,None]
y = np.zeros_like(x)
z = y
grid = np.concatenate((x,y,z), axis=1).T
```

### Wu and Yang

```python
with open('wuyang.json', encoding='utf-8') as F:
    wuyang = json.loads(F.read())
 
df = pd.DataFrame.from_dict(wuyang, orient='index')
df.T
```

```python
fig, ax = plt.subplots(1, 5, figsize=(20,5))
for i in range(5):
    ax[i]
    ax[i].plot(x.T[0,:], wuyang["vxc"][i], label=df.loc["basis"][i])
    ax[i].text(0.1,-9, s=f"Homo Value:\n {df.loc['occ_eigs'][i][4]:6f}")
    ax[i].legend()
    ax[i].set_ylim(-14,1)
    ax[i].set_xlim(0.001,9)
    ax[i].set_xscale('log')
    
fig.suptitle("Neon: WuYang Method")
```
